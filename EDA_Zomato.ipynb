***obs:*** huge majority of transactions are from India

```python
def pie_labels(labels, vals):
  labels_op = []
  vals_op = []
  rest=0
  for (i,val) in enumerate(vals):
    if val/sum(vals) > 0.01:
      labels_op.append(labels[i])
      vals_op.append(val)
    else:
      rest+=val

  labels_op.append('rest')
  vals_op.append(rest)

  return labels_op, vals_op
```

```python
country_names = df.Country.value_counts().index
country_vals = df.Country.value_counts().values

labels, vals = pie_labels(country_names, country_vals)

plt.pie(vals, labels=labels, autopct='%0.1f%%')

plt.show()
```

***obs:*** almost 9 out 10 transactions are from India

##***Obj5:*** most preferred cuisine

```python
cuisines_dict = {}

for lst in df['Cuisines']:
  temp_lst = lst.split(', ')
  for cuisine in temp_lst:
    if cuisine not in cuisines_dict:
      cuisines_dict[cuisine] = 1
    else:
      cuisines_dict[cuisine]+=1


sorted_cuisines = sorted(cuisines_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
cuisines_name = [item[0] for item in sorted_cuisines]
cuisines_count = [item[1] for item in sorted_cuisines]
```

```python
plt.bar(cuisines_name[:10], cuisines_count[:10])
plt.xticks(rotation=45)
plt.show()
```

```python
labels = cuisines_name[:5]
wedges = cuisines_count[:5]
plt.pie(wedges, labels=labels, autopct='%1.0f%%')
plt.show()
```

***obs:*** north indian is the most preferred cuisine
